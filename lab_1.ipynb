```python
# Коротко об этом датасете
# Cs:Go. Профессиональные матчи. Человек каждые 20 секунд фоткал каждый раунд и вытаскивал из него данные сколько секунд до конца раунда,
# сколько живых, с каким хп, сколько осталось в сумме у каждой команды, какое оружие и т.д.
# так же есть уже решенный исход раунда (кто выиграл)

# для классификации: предсказать кто победит
# для регрессии: сколько живых игроков у кт
```

```python
# кто победит?
# первое, что приходит в голову это посмотреть зависимость от кол-ва живых игроков

import seaborn as sns
from matplotlib import pyplot as plt

fig, ax = plt.subplots(ncols=2, figsize=(20, 5))
sns.violinplot(data=dSet, y='ct_players_alive', x='round_winner', ax=ax[0])
sns.violinplot(data=dSet, y='t_players_alive', x='round_winner', ax=ax[1])
```

```python
# сначала левый график, где кол-во живых контр-террористов (кт):
# если кт не впятером, то они с большей вероятностью проигрывают
# когда кт 5 человек, они выигрывают примерно в 3 раза чаще
# когда кт 4 человека, выигрывают примерно в 2 раза чаще
# 3 человека - примерно 50/50
# 2 человека - больше проигрывают
# и 1 - проигрывают

# на правом графике, где кол-во живых террористов, получеам примерно то же самое, но в обратном виде
# можно с уверенностью сказать, что исход раунда зависит от живых игроков в обеих командах
```

```python
# Далее рассмотрим два графика исхода раунда и кол-во денег у обеих команд
# команда чуть чаще выигрывает, когда у нее много денег
# это тоже можно использовать для предсказания

sns.catplot(data=dSet, x="round_winner", y="ct_money", height=10)
```

```python
sns.catplot(data=dSet, x="round_winner", y="t_money", height=10)
```

```python
# теперь время до конца раунда и победы
# когда остается меньше 50 секунд, чаще побеждают террористы
# также можно использовать для обучения
sns.catplot(data=dSet, x="round_winner", y="time_left", height=10)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.violinplot(data=dSet, y='t_players_alive', x='ct_players_alive', ax=ax)
# когда сколько игроков в обеих командах живые?
# когда в живых 5 кт, скорее всего живо 5 т
# когда в живых 4 кт, скорее всего у т живых 4-5
# дальше нет какой то особой свзяи, т.е. например у кт 2 живых, т может быть 2-4
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.scatterplot(data=dSet, x="ct_players_alive", y="time_left", ax=ax)

# этот график показывает в какой момент времени сколько живых игроков
# логично, что в начале раунда все 5 игроков живы. Смерти начинаются тогда, когда до конца раунда остается примерно 110 секунд
```

```python
sns.catplot(data=dSet, x="ct_players_alive", y="time_left", height=10)
```

```python
# далее построим корреляциб Пирсона. Если простыми словами, то она показывает линейную зависимость между двумя признаками
# чем ближе число по модулю к единице, тем больше связь
dSet.corr(method='pearson').style.background_gradient(cmap='coolwarm')
```

```python
# из корреляции видно, что самая сильная зависимость наблюдается между кол-вом живых игрогов за кт и их суммарным здоровьем.
# так же есть зависимость от времени до конца раунда и от кол-ва живых противников - террористов и их суммарного здоровья
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.violinplot(data=dSet, y='ct_health', x='ct_players_alive', ax=ax)
```

```python
# это прослеживается и в скрипичных графиках: чем меньше здоровья, тем меньше вживых игроков
```
